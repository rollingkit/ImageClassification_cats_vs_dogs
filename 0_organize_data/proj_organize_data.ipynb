```python
import os, shutil
```

```python
path = 'train/'
```

Create two directories 'cat' and 'dog' to move the respective files into. 
Then go through the names of all the files and if they begin with 'cat', move them to the 'cat' directory, and if they begin with 'dog', move them to 'dog' directory. 

The reason why 'if' statement is used for both cases and not just use 'else' for one, is to make sure there are no files or images from a third class or files that are named differently. 

```python
cat_folder = 'cat'
dog_folder = 'dog'
os.mkdir(os.path.join(path, cat_folder))
os.mkdir(os.path.join(path, dog_folder))
```

```python
for f in os.listdir(path):
  if f.split('.')[0] == 'cat':
    shutil.move(os.path.join(path, f), os.path.join(path, cat_folder))
```
